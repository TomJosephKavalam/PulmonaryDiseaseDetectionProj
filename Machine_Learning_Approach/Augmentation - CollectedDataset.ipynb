Augmentation

```python
sr=4000
```

```python
def aug(y):
    y_t_shift = t_shift(y)
    y_p_shift = pitch_shift(y)
    y_s_change = speed_change(y)
    return y_t_shift, y_p_shift, y_s_change
```

```python
def t_shift(y):
    y_shift_r = y[:-2*sr]
    return y_shift_r
```

```python
def pitch_shift(y):
    return lib.effects.pitch_shift(y, sr=sr, bins_per_octave=12, n_steps=4)
```

```python
def speed_change(y):
    return lib.effects.time_stretch(y, rate=1.2)
    
```

```python
y, sr=lib.load("CollectedDataset/Asth/3_AAL_W_A.wav", sr=sr)
y_t, y_p, y_s = aug(y)
display(y_t, y_p, y_s)
```

```python
for filename in glob.glob('CollectedDataset\Bron\*.wav'):
    y,sr=lib.load(filename, sr=4000)
    y_t, y_p, y_s = aug(y)
    file=filename.split("\\")[2].split(".")[0]
    
    print(file)
    sf.write("CollectedDataset\Bron_Aug\ "+ file+".wav", y, sr)
    sf.write("CollectedDataset\Bron_Aug\ "+ file+"_time_shift.wav", y_t, sr)
    sf.write("CollectedDataset\Bron_Aug\ "+ file+"_pitch_shift.wav", y_p, sr)
    sf.write("CollectedDataset\Bron_Aug\ "+ file+"_speed_change.wav", y_s, sr)
```
