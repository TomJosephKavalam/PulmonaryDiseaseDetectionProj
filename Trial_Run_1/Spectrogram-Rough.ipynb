```python
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
import librosa as lbr
import time
import scipy 
import pandas as pd
from os import walk
import glob
import librosa.display

# For plotting headlessly
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
```

for h in glob.glob("Audio files/Healthy/*.wav"):
    y, sr=lbr.load(h)
    print(h, 'Spectrogram/Healthy/'+h.split('\\')[1].split('.')[0]+'.png')
    plt.specgram(y, Fs=sr, NFFT=5000, noverlap=400, cmap='jet_r')
    plt.colorbar()
    plt.show()
    plt.savefig('Spectrogram/Healthy/'+h.split('\\')[1].split('.')[0]+'.png')

```python
y, sr = lbr.load("Audio files/Healthy/102_1b1_Ar_sc_Meditron.wav")

window_size = 1024
window = np.hanning(window_size)
stft  = lbr.core.spectrum.stft(y, n_fft=window_size, hop_length=512, window=window)
out = 2 * np.abs(stft) / np.sum(window)

fig = plt.Figure()
canvas = FigureCanvas(fig)
ax = fig.add_subplot(111)
p = lbr.display.specshow(lbr.amplitude_to_db(out, ref=np.max), ax=ax, y_axis='log', x_axis='time')
fig.savefig('spec.png')
np.savetxt('stftc.txt', stft, delimiter=',')
```
