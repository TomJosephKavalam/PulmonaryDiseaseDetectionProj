```python
current_dir=r"D:/Deep Learning/Pulmonary Disease - 2/AudioFiles2"
filenames=[]
for filename in os.listdir(current_dir):
    if filename.endswith('.txt'):
                print(filename)
                filenames.append(filename)
```

```python
j=0
for filename in filenames:
    txt=pd.read_csv('AudioFiles2/'+filename, sep="\t", header=None)#header=['Start', 'End', 'Wheeze', 'Crackle'])
    n=w=c=b=0
    print(filename)
    for i in txt.index:
            if txt[2][i]==0:
                if txt[3][i]==0:
                    n=n+1
                elif txt[3][i]==1:
                    w=w+1
            if txt[2][i]==1:
                if txt[3][i]==0:
                    c=c+1
                elif txt[3][i]==1:
                    b=b+1
    df['Patient_ID'][j]=int(filename.split('_')[0])
    df['Recording_index'][j]=filename.split('_')[1]
    df['Chest_Location'][j]=filename.split('_')[2]
    df['Acquisition_Mode'][j]=filename.split('_')[3]
    df['Recording_equipment'][j]=filename.split('_')[4].split('.')[0]
    df['Respiratory_cycles'][j]=n+c+w+b
    df['Normal_cycles'][j]=n
    df['Wheeze_cycles'][j]=w
    df['Crackle_cycles'][j]=c
    df['Both'][j]=b
    df['Filename'][j]=filename
    df['Train/Test'][j]=splitd[1][j]
    j=j+1
```

```python
for i in df.index:
    #print(patient_id1)
    #print(df[df['Patient_ID']==patient_id1]['Disease'])
    d=diag[1][diag[0]==df.loc[i]['Patient_ID']].tolist()
    df['Disease'][i]=d[0]
```
